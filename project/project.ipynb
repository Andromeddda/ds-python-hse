### Коэффициент Пирсона и его значимость

Проверим значимость наибольших значений кореляции:

* Session_Duration - Calories_Burned

        Гипотеза (H1): существует взаимосвязь между длительностью тренировки и количество сжигаемых калорий

* Session_Duration - Experience_Level

        Гипотеза (H1): существует взаимосвязь между опытом спортсмена и длительностью тренировки 

* Workout_Frequency - Experience_Level

        Гипотеза (H1): существует взаимосвязь между опытом спортсмена и частотой тренировок

* Calories_Burned - Experience_Level

        Гипотеза (H1): существует взаимосвязь между опытом спортсмена и количеством сжигаемых калорий

```python
cal = "Calories_Burned"
dur = "Session_Duration"
freq = "Workout_Frequency"
exp = "Experience_Level"

corr_pairs = [[dur, cal], [dur, exp], [freq, exp], [cal, exp]]

# Написать отчет о корреляции и отвергунть/принять гипотезы
def corrReport(A, B, coeff, p_val):
    # Вывести значения
    print("{} x {}\ncoefficient: {}]\np-value: {}".format(A, B, coeff, p_val))

    # Отвергнуть/принять гипотезы
    if (p_val < 0.05 and coeff > 0.5):
        print("Взаимосвязь между {} и {} присутствует на уровне значимости p < 0.05.\nH1 принимается, H0 отвергается\n\n".format(A, B))
    else:
        print("Взаимосвязь между {} и {} отсутствует.\nH0 принимается, H1 отвергается\n\n".format(A, B))

for A, B in corr_pairs:
    pearson = stats.pearsonr(df[A], df[B])
    corrReport(A, B, pearson[0], pearson[1])
```

Все коэффициенты оказались значимы даже на уровне p < 0.001

### Коэффициенты Спирмена

```python
for A, B in corr_pairs:
    spearman = stats.spearmanr(df[A], df[B])
    corrReport(A, B, spearman[0], spearman[1])
```

### Коэффициенты Кендалла

```python
for A, B in corr_pairs:
    kendall = stats.kendalltau(df[A], df[B])
    corrReport(A, B, kendall[0], kendall[1])
```

### Модель для количества сжигаемых калорий

**Гипотеза (H1)**: линейная регрессионная модель может предсказать количество сжигаемых калорий с точностью не менее 90% на уровне значимости F < 0.05

```python
import statsmodels.api as sm

# Добавим категориальыне столбцы как int
df['gender_male'] = (df['Gender'] == 'Male').astype(int)
df['type_strength'] = (df['Workout_Type'] == 'Strength').astype(int)

# метрические предикторы 
calories_metrical_predictors = ['Session_Duration', 'Water_Intake', 'Avg_BPM']

# категориальные предикторы
calories_categorial_predictors = ['type_strength', 'gender_male']

# все предикторы
calories_predictors = calories_metrical_predictors + calories_categorial_predictors

# убираем NULL
df1 = df.dropna(subset=calories_predictors + ['Calories_Burned'])

# Модель
x = df1[calories_predictors]
y = df1['Calories_Burned']
x_const = sm.add_constant(x)
calories_model = sm.OLS(y, x_const).fit()

print(calories_model.summary())
```

```python
print('R^2 = {:f}'.format(calories_model.rsquared))
print('F-test significance = {:f}'.format(calories_model.f_pvalue))
```

H1 принимается, H0 отвергается.

Модель значима и предсказывает 95.5% изменений.

В ходе построения модели подтвердилось, что тип тренировки не значим для сжигания калорий.

Также выяснилось, что потребление воды не значимо для сжигания калорий.

### Модель для процента подкожного жира

**Гипотеза (H1)**: линейная регрессионная модель может предсказать процент подкожного жира с точностью не менее 90% на уровне значимости F < 0.05

В этой модели заменим предиктор средней частоты сердцебиения на частоту тренировок.

```python
from scipy.stats.mstats import zscore

# метрические предикторы 
fat_metrical_predictors = ['Session_Duration', 'Water_Intake', 'Workout_Frequency']

# категориальные предикторы
fat_categorial_predictors = ['type_strength', 'gender_male']

# все предикторы
fat_predictors = fat_metrical_predictors + fat_categorial_predictors

# убираем NULL
df1 = df.dropna(subset=fat_predictors + ['Fat_Percentage'])

# Модель
x = df1[fat_predictors]
y = df1['Fat_Percentage']
fat_model = sm.OLS(zscore(y), zscore(x)).fit()

print(fat_model.summary())
```

```python
print('R^2 = {:f}'.format(fat_model.rsquared))
print('F-test significance = {:f}'.format(fat_model.f_pvalue))
```

```python
df.plot('Water_Intake', 'Fat_Percentage', kind='scatter')
```

H0 принимается, H1 отвергается.

Модель значима, но предсказывает лишь 59.5% изменений.

### Модель для индекса массы тела

В данной модели, очевидно, не следует использовать в качестве предикторов вес и рост, поскольку индекс массы тела напрямую рассчитывается через их значения.

**Гипотеза (H1)**: линейная регрессионная модель может предсказать индекс массы тела с точностью не менее 90% на уровне значимости F < 0.05

```python
# метрические предикторы 
bmi_metrical_predictors = ['Water_Intake']

# категориальные предикторы
bmi_categorial_predictors = ['type_strength', 'gender_male']

# все предикторы
bmi_predictors = bmi_metrical_predictors + bmi_categorial_predictors

# убираем NULL
df1 = df.dropna(subset=bmi_predictors + ['Fat_Percentage'])

# Модель
x = df1[bmi_predictors]
y = df1['Fat_Percentage']
bmi_model = sm.OLS(zscore(y), zscore(x)).fit()

print(bmi_model.summary())
```

```python
print('R^2 = {:f}'.format(bmi_model.rsquared))
print('F-test significance = {:f}'.format(bmi_model.f_pvalue))
```

H0 принимается, H1 отвергается.

## Бинарная логистическая регрессия

### Предваритальный анализ связи с категориальными предикторами

```python
df['is_pro'] = (df['Experience_Level'] == 3).astype(int)

def AgeCategory(age):
    if (age < 30):
        return "Young adult"
    if (age < 45):
        return "Middle age"
    return "Old-aged"

df['age_category'] = df['Age'].apply(AgeCategory)

categorial = ['age_category', 'Gender']

for c in categorial:
    ct = pd.crosstab(df['is_pro'], df[c])
    ct.plot.bar()
```

### Предварительный анализ связи с метрическими предикторами

```python
metrical = ['Session_Duration', 'Water_Intake', 'Workout_Frequency', 'Avg_BPM']

for m in metrical:
    plt.figure(figsize=(8,5))
    plt.title('{} vs. {}'.format(m, 'is_pro'), fontsize=16)
    plt.scatter(x=df[m], y=df['is_pro'], color='blue',edgecolor='k')
    plt.grid(True)
    plt.xlabel(m, fontsize=14)
    plt.ylabel('is_pro', fontsize=14)
    plt.show()
```

### Построение модели

```python
df['is_old'] = (df['Age'] > 45).astype(int)


x = df[metrical + ['gender_male', 'is_old']]
x_const = sm.add_constant(x)
y = df['is_pro']

pro_model = sm.Logit(y, x_const).fit()

print(pro_model.summary())
```

### Значимость модели

```python
print('R^2 = {:f}'.format(pro_model.prsquared))
print('Model significance = {:f}'.format(pro_model.llr_pvalue))
```

```python
l = []
for (param, index) in zip(pro_model.params[1:], pro_model.params.index[1:]):
    l.append('{:.2f}*{}'.format(param, index))
    
print('logit(p) = {:2f} + {}'.format(pro_model.params[0], ' + '.join(l)))
```

## Кластеризация

```python
from sklearn.cluster import KMeans, DBSCAN

# Выберем 5 переменных, которые слабо кореллируют
cols = ['Age', 'BMI', 'Workout_Frequency', 'Avg_BPM', 'Resting_BPM']

df[cols].corr()
```

```python
df0 = pd.DataFrame()

# Стандартизированные столбцы
z_cols = []
for i in cols:
  z_cols.append('z_' + i)
  df0["z_" + i] = (df[i] - df[i].mean())/df[i].std() #стандартизируем значения переменных
  df0[i] = df[i]

df0
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
# Применим несколько алгоритмов иерархической кластеризации

# Близлежащий сосед
plt.figure(figsize=(15, 5))
dn = dendrogram(linkage(df0[z_cols], 'ward'))
```

```python
# Найдем количество кластеров через индекс Калински-Харабаша
from sklearn.metrics import calinski_harabasz_score

Z = linkage(df0[z_cols], 'ward')

for k in range (2, 10):
    labels = fcluster(Z, t=k, criterion='maxclust')
    print('Число кластеров: {}, индекс: {}'.format(k, calinski_harabasz_score(df0[z_cols], labels)))
```

```python
from yellowbrick.cluster import KElbowVisualizer

inertia = []
X = df0[z_cols]

k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Plot')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(k_range)

model = KMeans(random_state=1)
visualizer = KElbowVisualizer(model, k=(2,10), timings=False)

visualizer.fit(X)
visualizer.show()
plt.show()
```

Выведем все ссылки, содержащие на пути 'feeds' - это категории. Так можно будет со страницы каждой категории собрать по несколько видео.

```python
urls = [a['href'] for a in soup.find_all("a", href=True) if 'feeds' in a['href']]

for i in range(0, len(urls)):
    if urls[i][0] == '/':
        urls[i] = root + urls[i]
        
urls
```

```python
# Обходим страницы 10 категорий и собираем с них все ссылки на видео

print('Total categories: {}'.format(len(urls)))

video_urls = []

import re
def is_video(url):
    if ('person' in url or 'live' in url):
        return False
    return re.match(r"(/video/.+)|(.*rutube.ru/video/.+)", url)

for feed in urls[:10]:
    page = requests.get(feed)
    soup0 = BeautifulSoup(page.text, 'html.parser')
    new_urls = [a['href'] for a in soup0.find_all("a", href=True) if is_video(a['href'])]
    video_urls += new_urls
```

```python
print('Total videos: {}'.format(len(video_urls)))

video_urls
```

```python
# добавляем корень к неполным путям
for i in range(len(video_urls)):
    if (video_urls[i][0] == '/'):
        video_urls[i] = root + video_urls[i]

video_urls
```

```python
# рассмотрим видео по первой ссылке

new_0 = requests.get(video_urls[-1])
soup0 = BeautifulSoup(new_0.text, 'html.parser')
soup0.findAll('meta')
```

```python
# теперь напишем функцию сбора информации

def find(soup, what):
    try:
        return soup.findAll('meta', what)[-1].attrs['content']
    except:
        return ""

def GetInfo(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    author = find(soup, {'property': 'ya:ovs:login'})
    title = find(soup, {'name': 'shorttitle'})
    release_date = find(soup, {'property': 'og:video:release_date'})
    upload_date = find(soup, {'property': 'ya:ovs:upload_date'})
    category = find(soup, {'property': 'ya:ovs:category'})
    
    duration = find(soup, {'property': 'og:video:duration'})
    is_official = find(soup, {'property': 'ya:ovs:is_official'})
    adult = find(soup, {'property': 'ya:ovs:adult'})
    width = find(soup, {'property': 'og:video:width'})
    height = find(soup, {'property': 'og:video:height'})
    views = find(soup, {'property': 'ya:ovs:views_total'})
   

    return author, title, release_date, upload_date, duration, category, is_official, adult, width, height, views
```

```python
videos = []

for v in video_urls[:200:2]:
    videos.append(GetInfo(v))
```
